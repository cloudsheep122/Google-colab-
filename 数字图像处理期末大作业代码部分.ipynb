```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def convolution(image, kernel):
    # 使用OpenCV中的filter2D函数进行卷积
    result = cv2.filter2D(image, -1, kernel)
    return result

def visualize_result(image, result):
    plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.subplot(122), plt.imshow(result, cmap='gray'), plt.title('Convolution Result')
    plt.show()

def save_result(result, output_path):
    cv2.imwrite(output_path, result)

# 读取图像
image_path = '/content/image.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 定义卷积核，这里以一个简单的3x3均值滤波器为例
kernel = np.ones((3, 3), np.float32) / 9

# 进行卷积运算
result = convolution(image, kernel)

# 可视化结果
visualize_result(image, result)

# 保存结果图像
output_path = '/content/result.png'
save_result(result, output_path)
```

```python
def median_filter(image, salt_and_pepper_flag):
    if salt_and_pepper_flag:
        # 如果有椒盐噪声，则执行中值滤波
        result = cv2.medianBlur(image, 3)
        return result
    else:
        # 如果没有椒盐噪声，则返回原图像
        return image

def visualize_result(image, result):
    plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.subplot(122), plt.imshow(result, cmap='gray'), plt.title('Median Filter Result')
    plt.show()

def save_result(result, output_path):
    cv2.imwrite(output_path, result)

# 读取图像
image_path = '/content/image.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 假设salt_and_pepper_flag为True表示有椒盐噪声，为False表示没有
salt_and_pepper_flag = True

# 执行中值滤波
result = median_filter(image, salt_and_pepper_flag)

# 可视化结果
visualize_result(image, result)

# 保存结果图像
output_path = '/content/result2.png'  # 替换成你想要保存的路径
save_result(result, output_path)
```

```python

def frequency_domain_filter(image, filter_type):
    # 进行DCT变换
    dct_image = cv2.dct(np.float32(image))

    # 定义频域滤波器，这里以一个简单的低通或高通滤波器为例
    rows, cols = image.shape
    crow, ccol = rows // 2, cols // 2

    # 创建滤波器
    mask = np.zeros_like(dct_image)
    if filter_type == 'lowpass':
        mask[crow-30:crow+30, ccol-30:ccol+30] = 1
    elif filter_type == 'highpass':
        mask[0:rows, 0:cols] = 1
        mask[crow-30:crow+30, ccol-30:ccol+30] = 0

    # 应用滤波器
    filtered_dct = dct_image * mask

    # 进行IDCT反变换
    filtered_image = cv2.idct(filtered_dct)

    # 为了显示，将结果映射到0-255范围
    filtered_image = np.uint8(np.clip(filtered_image, 0, 255))

    return filtered_image

def visualize_result(image, result):
    plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.subplot(122), plt.imshow(result, cmap='gray'), plt.title('Frequency Domain Filter Result')
    plt.show()

def save_result(result, output_path):
    cv2.imwrite(output_path, result)

# 读取图像
image_path = '/content/image3.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 定义频域滤波器类型，可选 'lowpass' 或 'highpass'
filter_type = 'lowpass'

# 执行频域滤波
result = frequency_domain_filter(image, filter_type)

# 可视化结果
visualize_result(image, result)

# 保存结果图像
output_path = '/content/result3.png'
save_result(result, output_path)
```

```python

def histogram_equalization(image):
    # 使用OpenCV中的直方图均衡化函数
    equalized_image = cv2.equalizeHist(image)
    return equalized_image

def histogram_matching(image, reference_histogram):
    # 计算输入图像的直方图
    hist_image, _ = np.histogram(image.flatten(), bins=256, range=[0,256])

    # 计算累积直方图
    cum_hist_image = np.cumsum(hist_image) / float(np.sum(hist_image))

    # 计算参考直方图的累积直方图
    cum_ref_hist = np.cumsum(reference_histogram) / float(np.sum(reference_histogram))

    # 映射表
    mapping_table = np.interp(cum_hist_image, cum_ref_hist, range(256))

    # 应用映射表
    matched_image = np.floor(np.interp(image, range(256), mapping_table)).astype('uint8')

    return matched_image

def visualize_result(image, result, title):
    plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
    plt.subplot(122), plt.imshow(result, cmap='gray'), plt.title(title)
    plt.show()

def save_result(result, output_path):
    cv2.imwrite(output_path, result)

# 读取图像
image_path = '/content/image6.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 执行直方图均衡化
equalized_image = histogram_equalization(image)
visualize_result(image, equalized_image, 'Histogram Equalization Result')

# 定义参考直方图，这里使用均匀分布的直方图作为参考
reference_histogram = np.ones(256) * (1 / 256)

# 执行直方图规定化
matched_image = histogram_matching(image, reference_histogram)
visualize_result(image, matched_image, 'Histogram Matching Result')

# 保存结果图像
output_path_equalized = '/content/result5.png'
output_path_matched = '/content/result6.png'
save_result(equalized_image, output_path_equalized)
save_result(matched_image, output_path_matched)
```

```python

def meanshift_segmentation(image):
    # 将图像转换为Lab颜色空间
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

    # 提取L通道和a通道
    l_channel, a_channel, _ = cv2.split(lab_image)

    # 将L和a通道合并为特征矩阵
    feature_matrix = np.column_stack((l_channel.flatten().astype(np.float32), a_channel.flatten().astype(np.float32)))


    # 设置Meanshift参数
    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    _, labels, centers = cv2.kmeans(feature_matrix, 2, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    # 根据标签生成分割结果
    segmentation_result = labels.reshape(image.shape[:2])

    return segmentation_result

def visualize_result(image, segmentation_result):
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(segmentation_result, cmap='gray'), plt.title('Meanshift Segmentation Result')
    plt.show()

# 读取彩色图像
image_path = '/content/image7.jpg'
image = cv2.imread(image_path)

# 执行Meanshift图像分割
segmentation_result = meanshift_segmentation(image)

# 可视化结果
visualize_result(image, segmentation_result)

# 保存结果图像
output_path = '/content/result7.png'
cv2.imwrite(output_path, segmentation_result)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from google.colab.patches import cv2_imshow

def block_matching_disparity(left_image_path, right_image_path, window_size=5, max_disparity=16):
    # 读取左右矫正后的图像
    left_image = cv2.imread(left_image_path, cv2.IMREAD_GRAYSCALE)
    right_image = cv2.imread(right_image_path, cv2.IMREAD_GRAYSCALE)

    if left_image is None or right_image is None:
        raise ValueError("Error reading images. Make sure the image paths are correct.")

    # 获取图像尺寸
    height, width = left_image.shape[:2]

    # 初始化视差图
    disparity_map = np.zeros_like(left_image, dtype=np.uint8)

    # 设置窗口半径
    half_window = window_size // 2

    for y in range(half_window, height - half_window):
        for x in range(half_window, width - half_window):
            # 定义当前窗口
            window_left = left_image[y - half_window:y + half_window + 1, x - half_window:x + half_window + 1]

            # 在右图像中搜索匹配窗口
            search_region = right_image[y - half_window:y + half_window + 1, x - max_disparity - half_window:x + half_window + 1]

            # 检查搜索区域是否足够大
            if search_region.shape[1] < window_size:
                continue

            # 执行模板匹配
            result = cv2.matchTemplate(search_region, window_left, cv2.TM_SQDIFF)
            min_val, _, _, min_loc = cv2.minMaxLoc(result)

            # 计算对应的视差
            disparity = x - (min_loc[0] + half_window)

            # 将视差值存储到视差图中
            disparity_map[y, x] = disparity

    return disparity_map, width, height

# 图像路径
left_image_path = '/content/image9.jpg'
right_image_path = '/content/image.jpg'

# 执行窗口匹配视差估计
disparity_map, width, height = block_matching_disparity(left_image_path, right_image_path)

# 显示视差图
cv2_imshow(disparity_map)

# 计算深度图
baseline = 1.0  # 相机基线，单位是米
focal_length = 500  # 相机焦距，单位是像素
depth_map = (baseline * focal_length) / (disparity_map + 1e-6)  # 防止除零错误

# 可视化深度图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.meshgrid(range(width), range(height))
ax.plot_surface(x, y, depth_map, cmap='viridis')
plt.show()
```
